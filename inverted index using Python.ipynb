```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
Stopwords = set(stop)
```

```python
docs = ['burgess-busterbrown.txt',
 'edgeworth-parents.txt',
 'blake-poems.txt']
```

```python
dict_global = {}

for file in docs:
    words = gutenberg.words(file)
    words = list(words)
    words = [word for word in words if word.isalpha()]
    words = [word for word in words if len(words)>1]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in Stopwords]
    dict=(nltk.FreqDist(words))
    for i in dict.keys():
      if i not in dict_global.keys():
        dict_global[i]=dict[i]
      else:
        dict_global[i]=dict_global[i]+dict[i]
dict_global
```

```python
import collections
od = collections.OrderedDict(sorted(dict_global.items()))
od
```

```python
len(od)
```

```python
postingsList = {}
for i in od.keys():
  postingsList[i] = {}
  postingsList[i]["freq"] = od[i]
  postingsList[i]["docs"] = []
  for file in docs: 
    words = gutenberg.words(file)
    words = list(words)
    words = [word for word in words if word.isalpha()]
    words = [word for word in words if len(words)>1]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in Stopwords]
    dict=(nltk.FreqDist(words))
    if i in dict.keys():
      postingsList[i]["docs"].append(docs.index(file))
  
```

```python
postingsList
```

```python
def query(qry):
  for i in qry.split():
    if i=='and':
      qry= qry.replace(i,'&')
    elif i=='or':
      qry=qry.replace(i,'|')
    elif i=='not':
      qry=qry.replace(i,"{0,1,2}-")
  for i in qry.split():
    if i.isalpha():
      qry=qry.replace(i,repr(set(postingsList[i]['docs'])))
  return qry
```

```python
qry=input("Enter the query :")
d=eval(query(qry))
if d=={}:
  print("No documents are found satisfying the query")
else:
  for i in d:
    print("Document",i,end=", ")
  print("found")
```
